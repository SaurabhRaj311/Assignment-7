MongoDB is preferred when:

**d) How does embedding vs referencing impact performance?**

**Embedding**
- Related data is stored inside the same document.
- Requires fewer database queries.
- Faster read performance.
- Best for one-to-one or one-to-few relationships.

Example: Customer information and shipping address stored in one document.

**Referencing**
- Related data is stored in separate documents using references.
- Requires additional queries to fetch related data.
- Better for large datasets and many-to-many relationships.
- Reduces data duplication.

Example: Orders stored separately with a reference to the customer ID.

##Question 2: Indexing, Sharding & Replication
a) How do indexes improve MongoDB query performance?
b) What is sharding? When is it required?
c) What is replication and how does it ensure high availability?
d) How does improper shard key selection impact performance?

**Answer**

**a) How do indexes improve MongoDB query performance?**
Indexes help MongoDB find data quickly without scanning every document in a collection. They create a searchable structure on one or more fields, making queries faster and improving overall database performance.

Advantages:
- Faster data retrieval
- Reduced query execution time
- Better performance for large datasets

**b) What is sharding? When is it required?**

Sharding is the process of dividing a large database into smaller parts called shards, which are stored across multiple servers.
Sharding is required when:
- The database becomes very large.
- A single server cannot handle the workload.
- High read/write traffic needs to be managed.
- Horizontal scaling is required.

**c) What is replication and how does it ensure high availability?**

Replication is the process of maintaining multiple copies of the same data on different servers.
It ensures high availability because:
- If the primary server fails, a secondary server automatically becomes the new primary.
- Data remains available with minimal downtime.
- It provides backup and improves data reliability.

**d) How does improper shard key selection impact performance?**

A shard key determines how data is distributed across shards.
If an improper shard key is chosen:
- Data may not be distributed evenly.
- One shard may receive most of the requests (hotspot).
- Some servers become overloaded while others remain underutilized.
- Query performance decreases and scalability is affected.

Therefore, selecting an appropriate shard key is essential for balanced data distribution and efficient performance.

##Question 5: Write MongoDB queries to:
1. Insert a new order
2. Find all orders above ₹5000
3. Update status to "Delivered" where order_date is older than 7 days
4. Delete cancelled orders


**Answers**

```python
#installing required library
!pip install mongomock
```

```python
##creating databse and collection
import mongomock
from datetime import datetime, timedelta

client = mongomock.MongoClient()

db = client["shopDB"]
orders = db["orders"]
```

```python
##inserting sample orders
orders.insert_many([
    {
        "order_id": 1,
        "customer_id": 101,
        "amount": 6500,
        "status": "Pending",
        "order_date": datetime.now() - timedelta(days=10)
    },
    {
        "order_id": 2,
        "customer_id": 102,
        "amount": 3000,
        "status": "Cancelled",
        "order_date": datetime.now() - timedelta(days=2)
    }
])
```

```python
##a) insert a new order
orders.insert_one({
    "order_id": 3,
    "customer_id": 103,
    "amount": 8000,
    "status": "Pending",
    "order_date": datetime.now()
})

print("New order inserted successfully.")
```

```python
##b) Find all orders above ₹5000
result = orders.find({"amount": {"$gt": 5000}})

for order in result:
    print(order)
```

```python
##c) Update status to "Delivered" where order_date is older than 7 days
orders.update_many(
    {
        "order_date": {
            "$lt": datetime.now() - timedelta(days=7)
        }
    },
    {
        "$set": {"status": "Delivered"}
    }
)

print("Orders updated successfully.")
```

```python
##d) delete cancelled orders
orders.delete_many({"status": "Cancelled"})

print("Cancelled orders deleted successfully.")
```

## Question 8: Suppose:
● 5% of customers churn
● 70% of churned customers had declining order frequency
● 15% of non-churned customers had declining order frequency
If a customer shows declining frequency, compute probability that they will churn

**Answer**

```python
# Given probabilities
P_churn = 0.05                 # P(C)
P_decline_given_churn = 0.70   # P(D|C)
P_decline_given_not_churn = 0.15   # P(D|~C)

# Probability of not churning
P_not_churn = 1 - P_churn

# Calculate P(D)
P_decline = (
    P_decline_given_churn * P_churn
    + P_decline_given_not_churn * P_not_churn
)

# Apply Bayes' Theorem
P_churn_given_decline = (
    P_decline_given_churn * P_churn
) / P_decline

print("Probability of Churn =", round(P_churn_given_decline, 4))
print("Percentage =", round(P_churn_given_decline * 100, 2), "%")
```

##Question 9:
 Explain:
1. Why CLT allows us to use normal approximation
2. Why it is important in A/B testing
3. What happens if sample size is very small?

**Answers**

**a) Why does CLT allow us to use the normal approximation?**

- The Central Limit Theorem (CLT) states that as the sample size becomes sufficiently large (typically 30 or more), the distribution of the sample mean approaches a normal distribution, even if the original population is not normally distributed.

This makes it possible to use normal distribution methods for statistical analysis and hypothesis testing.

**b) Why is it important in A/B Testing?**

CLT is important in A/B testing because:
- It allows comparison of average results between two groups.
- It makes statistical tests (such as the Z-test) valid for large samples.
- It helps determine whether the observed difference is statistically significant or due to random chance.
- It supports data-driven business decisions.

**c) What happens if the sample size is very small?**

If the sample size is very small:
- The sample mean may not follow a normal distribution.
- Statistical results become less reliable.
- Estimates may have higher variability.
- Conclusions can be inaccurate or misleading.

If the population is not normally distributed and the sample size is small, alternative methods such as the t-test are generally preferred.

##Question 10:The company claims:
“Average monthly spending is ₹5000.”
1. How would you statistically test this claim?
2. Define null and alternative hypothesis
3. What type of error is more dangerous in business — Type I or Type II?


**Answers**


```python
#a) how would you statistically test this claim
#We can use a One-Sample t-test to test whether the average monthly spending is equal to ₹5000.
from scipy.stats import ttest_1samp

# Sample monthly spending data
spending = [5200, 4800, 5100, 4950, 5300, 5000, 4900, 5050, 5150, 4850]

# Perform one-sample t-test
t_statistic, p_value = ttest_1samp(spending, 5000)

print("T-statistic:", round(t_statistic, 4))
print("P-value:", round(p_value, 4))

if p_value < 0.05:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
#b) null and alternative hypothesis
# Hypothesis representation in Python

mu_0 = 5000  # claimed average spending

print("NULL HYPOTHESIS (H0):")
print("Average monthly spending is equal to 5000")
print("H0: mu = ", mu_0)

print("\nALTERNATIVE HYPOTHESIS (H1):")
print("Average monthly spending is NOT equal to 5000")
print("H1: mu != ", mu_0)
```

**c) What type of error is more dangerous in business — Type I or Type II?**

**Type I Error**

Rejecting a true null hypothesis.

Example: Concluding that the average spending is not ₹5000 when it actually is.

**Type II Error**

Failing to reject a false null hypothesis.

Example: Accepting that the average spending is ₹5000 when it is actually
different.

**Which is more dangerous?**

In many business situations, Type II Error is more dangerous because it may cause the company to overlook real problems or miss important opportunities. For example, failing to detect a decline in customer spending could delay corrective action and lead to financial losses.
